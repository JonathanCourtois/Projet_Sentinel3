# Handling a Sentinel-3 OLCI product with SNAP in Python (Tutorial 3)

## Subsetting a Sentinel3 data product

```python
from snappy import ProductIO
from snappy import jpy
from snappy import GPF

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
```

Chargement de la première image test depuis l'ordinateur.

```python
file_path = '/home/courtois/jupyter-sentinel/S3A_OL_2_WFR____20190322T095802_20190322T100102_20190323T184908_0179_042_350_2160_MAR_O_NT_002.SEN3/xfdumanifest.xml'
product = ProductIO.readProduct(file_path)
HashMap = jpy.get_type('java.util.HashMap')  
parameters = HashMap()
parameters.put('targetResolution',300)
result = product
```

```python
SubsetOp = jpy.get_type('org.esa.snap.core.gpf.common.SubsetOp')
WKTReader =jpy.get_type('com.vividsolutions.jts.io.WKTReader')
```

On definit la zone d'intéret grâce à SNAP.
le recupération des coordonées de la zone s'effectue grâce à :
clic droit sur la zone -> WKT from Geometry

<img src="AOI.png">

Defntion du polygone

```python
#marseille#
#wkt = 'POLYGON((3.842688570706235 43.96558560512889, 5.5955283808025404 43.68763559200283,5.281623963550708 42.686759820995256, 3.55692959217631 42.96334035376373,3.842688570706235 43.96558560512889))'
#bordeau#
wkt = 'POLYGON ((-1.8272688312669267 46.21406553869655, -0.2835733912720454 46.063464013417146,-0.5003738814714196 45.0540256728009, -2.0163206748476883 45.206150300245845,-1.8272688312669267 46.21406553869655))'
geometry = WKTReader().read(wkt)
```

Once the geometry has been  properly defined, the subset operator can be invoked with its object constructor, and then it will be possible to call its methods that will allow to set the geometry object and the source product:
Une fois la géometrie proprement definit, on créé un sous ensemble de l'objet qu'il sera possible d'appeler pour modifier les réglages de l'objet.

```python
op = SubsetOp()
op.setSourceProduct(result)
op.setGeoRegion(geometry)
sub_product = op.getTargetProduct()
print("subset product 1 ready")
```

Now __sub_product__ is the new data product obtained from the source S-3 product. We are going to verify the result by showing the list of available bands, and visualizing one of them.


```python
list(result.getBandNames())
```

on observe une des bandes disponible:

la bande Oa06 a 560nm

```python
B6 = product.getBand('Oa06_reflectance')
Width = B6.getRasterWidth()
Height = B6.getRasterHeight()
print("Image full scale : Width:",Width," ; Height:",Height)
```

```python
B6_data = np.zeros(Width*Height, dtype = np.float32)
B6.readPixels(0,0,Width,Height,B6_data)
B6_data.shape = (Height,Width)
```

### affichage de l'image

```python
plt.figure(figsize=(8, 8))                 # adjusting the figure window size
fig = plt.imshow(B6_data, cmap = cm.gray)  #matplotlib settings for the current image
fig.axes.get_xaxis().set_visible(False)
fig.axes.get_yaxis().set_visible(False)
plt.show()
```

deffinition d'une nouvelle image correspondant uniquement au coordonées utiles de la première image

```python
sub_b6 = sub_product.getBand('Oa06_reflectance')
print(sub_b6)
print("band read")
width = sub_b6.getRasterWidth()
height = sub_b6.getRasterHeight()
print("subset size : ", width, height)

sub_b6_data = np.zeros(width*height, dtype = np.float32)
sub_b6.readPixels(0,0,width,height, sub_b6_data)
sub_b6_data.shape = height, width
print("reading data...")

plt.figure(1)
fig = plt.imshow(sub_b6_data, cmap = cm.gray)
fig.axes.get_xaxis().set_visible(False)
fig.axes.get_yaxis().set_visible(False)
plt.show()
```
